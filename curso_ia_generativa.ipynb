```python
import pandas as pd
import numpy as np

# carregar dataset
df = pd.read_excel('curso_ia_generativa.xlsx')

print("Dimensão do dataset:")
print(df.shape)

print("\nPrimeiras linhas:")
display(df.head())

print("\nInformações das variáveis:")
df.info()

print("\nEstatísticas descritivas:")
display(df.describe(include='all'))
```

```python
import pandas as pd
import numpy as np

# 1) distribuição de participantes por grupo
print("Participantes por grupo:")
print(df["grupo"].value_counts(dropna=False))

# 2) checar trilhas por tópico (t01 a t04)
trilha_cols = ["t01_trilha", "t02_trilha", "t03_trilha", "t04_trilha"]
print("\nValores únicos por coluna de trilha:")
for c in trilha_cols:
    print(f"{c}: {sorted(df[c].astype(str).unique())}")

# 3) validar score no grupo controle (deve ser -1)
score_cols = ["t01_score", "t02_score", "t03_score", "t04_score"]
ctrl = df[df["grupo"].str.lower() == "controle"]

print("\nValidação: scores do grupo Controle (min, max) — esperado (-1, -1):")
for c in score_cols:
    print(f"{c}: min={ctrl[c].min()}, max={ctrl[c].max()}")

# 4) validar que Experimental NÃO tem score -1 (ou tem poucos casos estranhos)
exp = df[df["grupo"].str.lower() == "experimental"]
print("\nValidação: quantos '-1' no grupo Experimental (esperado 0):")
for c in score_cols:
    print(f"{c}: {(exp[c] == -1).sum()}")

# 5) checar faixas das escalas (sanidade)
print("\nChecagens de faixa (sanidade):")
print("t01..t04 (acertos) min/max:", df[["t01","t02","t03","t04"]].min().min(), df[["t01","t02","t03","t04"]].max().max())
print("dif (Likert) min/max:", df[["t01_dif","t02_dif","t03_dif","t04_dif"]].min().min(), df[["t01_dif","t02_dif","t03_dif","t04_dif"]].max().max())
print("autoc (Likert) min/max:", df[["t01_autoc","t02_autoc","t03_autoc","t04_autoc"]].min().min(), df[["t01_autoc","t02_autoc","t03_autoc","t04_autoc"]].max().max())
print("nota_pre min/max:", df["nota_pre"].min(), df["nota_pre"].max())
print("nota_pos min/max:", df["nota_pos"].min(), df["nota_pos"].max())
```

```python
# ===============================
# BLOCO 3 — Variáveis derivadas
# ===============================

# médias cognitivas dos tópicos
nota_cols = ["t01_nota","t02_nota","t03_nota","t04_nota"]

df["cognitivo_mean"] = df[nota_cols].mean(axis=1)

# média metacognitiva (dificuldade)
dif_cols = ["t01_dif","t02_dif","t03_dif","t04_dif"]

df["metacognitivo_mean"] = df[dif_cols].mean(axis=1)

# média afetiva (autoconfiança)
auto_cols = ["t01_autoc","t02_autoc","t03_autoc","t04_autoc"]

df["afetivo_mean"] = df[auto_cols].mean(axis=1)

# UX geral
ux_cols = ["ux1","ux2","ux3","ux4","ux5","ux6","ux7"]

df["UX_mean"] = df[ux_cols].mean(axis=1)

print("Novas variáveis criadas:")
print(df[["cognitivo_mean","metacognitivo_mean","afetivo_mean","UX_mean"]].head())

print("\nEstatísticas gerais:")
display(df[["cognitivo_mean","metacognitivo_mean","afetivo_mean","UX_mean"]].describe())
```

```python
from scipy.stats import shapiro, levene, ttest_ind, mannwhitneyu
import numpy as np

# separar grupos
controle = df[df["grupo"]=="Controle"]
experimental = df[df["grupo"]=="Experimental"]

print("Tamanho dos grupos:")
print("Controle:", len(controle))
print("Experimental:", len(experimental))

# ======================
# PRE-TESTE
# ======================

pre_ctrl = controle["nota_pre"]
pre_exp = experimental["nota_pre"]

print("\n--- PRE-TESTE ---")

print("Controle:", pre_ctrl.values)
print("Experimental:", pre_exp.values)

# normalidade
w0,p0 = shapiro(pre_ctrl)
w1,p1 = shapiro(pre_exp)

print("\nShapiro Controle:", p0)
print("Shapiro Experimental:", p1)

# homogeneidade
statL,pL = levene(pre_ctrl,pre_exp)

print("\nLevene p:",pL)

# t-test
t,p = ttest_ind(pre_ctrl,pre_exp,equal_var=True)

print("\nT-test")
print("t:",t)
print("p:",p)

# ======================
# POS-TESTE
# ======================

pos_ctrl = controle["nota_pos"]
pos_exp = experimental["nota_pos"]

print("\n--- POS-TESTE ---")

print("Controle:", pos_ctrl.values)
print("Experimental:", pos_exp.values)

# normalidade
w0,p0 = shapiro(pos_ctrl)
w1,p1 = shapiro(pos_exp)

print("\nShapiro Controle:", p0)
print("Shapiro Experimental:", p1)

# homogeneidade
statL,pL = levene(pos_ctrl,pos_exp)

print("\nLevene p:",pL)

# t-test
t,p = ttest_ind(pos_ctrl,pos_exp,equal_var=True)

print("\nT-test")
print("t:",t)
print("p:",p)

# ======================
# GANHO DE APRENDIZAGEM
# ======================

df["ganho"] = df["nota_pos"] - df["nota_pre"]

ganho_ctrl = controle["nota_pos"] - controle["nota_pre"]
ganho_exp = experimental["nota_pos"] - experimental["nota_pre"]

print("\n--- GANHO ---")

print("Controle:", ganho_ctrl.values)
print("Experimental:", ganho_exp.values)

# t-test
t,p = ttest_ind(ganho_ctrl,ganho_exp,equal_var=True)

print("\nT-test ganho")
print("t:",t)
print("p:",p)
```

```python
from scipy.stats import mannwhitneyu
import numpy as np

# separar grupos
controle = df[df["grupo"]=="Controle"]
experimental = df[df["grupo"]=="Experimental"]

pre_ctrl = controle["nota_pre"]
pre_exp = experimental["nota_pre"]

pos_ctrl = controle["nota_pos"]
pos_exp = experimental["nota_pos"]

ganho_ctrl = pos_ctrl - pre_ctrl
ganho_exp = pos_exp - pre_exp

# ================================
# COHEN'S D
# ================================

def cohens_d(a,b):
    n1 = len(a)
    n2 = len(b)

    s1 = np.std(a, ddof=1)
    s2 = np.std(b, ddof=1)

    pooled = np.sqrt(((n1-1)*s1**2 + (n2-1)*s2**2)/(n1+n2-2))

    d = (np.mean(a)-np.mean(b))/pooled

    return d

print("\n======================")
print("COHEN'S D")
print("======================")

d_pre = cohens_d(pre_ctrl,pre_exp)
d_pos = cohens_d(pos_ctrl,pos_exp)
d_gain = cohens_d(ganho_ctrl,ganho_exp)

print("Pré-teste:",d_pre)
print("Pós-teste:",d_pos)
print("Ganho:",d_gain)

# interpretação

def interpret_d(d):

    d=abs(d)

    if d<0.2:
        return "efeito muito pequeno"
    elif d<0.5:
        return "efeito pequeno"
    elif d<0.8:
        return "efeito moderado"
    else:
        return "efeito grande"

print("\nInterpretação:")
print("Pré:",interpret_d(d_pre))
print("Pós:",interpret_d(d_pos))
print("Ganho:",interpret_d(d_gain))


# ================================
# GANHO NORMALIZADO (HAKE)
# ================================

print("\n======================")
print("GANHO NORMALIZADO (HAKE)")
print("======================")

hake_ctrl = (pos_ctrl.mean() - pre_ctrl.mean()) / (10 - pre_ctrl.mean())
hake_exp = (pos_exp.mean() - pre_exp.mean()) / (10 - pre_exp.mean())

print("Hake Controle:",hake_ctrl)
print("Hake Experimental:",hake_exp)

# classificação Hake

def interpret_hake(g):

    if g<0.3:
        return "ganho baixo"
    elif g<0.7:
        return "ganho médio"
    else:
        return "ganho alto"

print("\nClassificação:")
print("Controle:",interpret_hake(hake_ctrl))
print("Experimental:",interpret_hake(hake_exp))


# ================================
# MANN WHITNEY
# ================================

print("\n======================")
print("MANN-WHITNEY")
print("======================")

u_pre,p_pre = mannwhitneyu(pre_ctrl,pre_exp)
u_pos,p_pos = mannwhitneyu(pos_ctrl,pos_exp)
u_gain,p_gain = mannwhitneyu(ganho_ctrl,ganho_exp)

print("Pré-teste: U=",u_pre," p=",p_pre)
print("Pós-teste: U=",u_pos," p=",p_pos)
print("Ganho: U=",u_gain," p=",p_gain)
```
